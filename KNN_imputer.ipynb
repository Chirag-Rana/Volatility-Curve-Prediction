```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')

# Load data
data_path = '/kaggle/input/nk-iv-prediction'
test = pd.read_parquet(f'{data_path}/test_data.parquet')
sample_submission = pd.read_csv(f'{data_path}/sample_submission.csv')

target_columns = [col for col in sample_submission.columns if col != 'timestamp']

# Start with sample_submission structure
submission = sample_submission.copy()

# Copy non-missing values from test to submission
for col in target_columns:
    if col in test.columns:
        submission[col] = test[col].values
    else:
        submission[col] = np.nan  # Placeholder for missing columns

# Impute missing values using KNNImputer (on all target columns at once)
imputer = KNNImputer(n_neighbors=13)
imputed_values = imputer.fit_transform(submission[target_columns])
submission[target_columns] = imputed_values

# Clip values to a realistic IV range (1% to 300%)
submission[target_columns] = submission[target_columns].clip(0.01, 3.0)

# Save to CSV
submission.to_csv('submission.csv', index=False)
print("submission saved!")
```
